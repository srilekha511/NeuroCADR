```python
no_match = []
import numpy as np
#for i,v in enumerate(df2.iloc[:,0]):
#    sub = df3[df3.iloc[:,4].str.contains(v)] 
#    for j in sub.iloc[:,5]:
#        res = np.where(symptoms == j)
 #       df2.iloc[i, (res[0]+2)] = 1

for i,v in enumerate(df2.iloc[:,0]):
    if v in df3.iloc[:,4].unique():
        sub = df3[df3.iloc[:,4].str.contains(v)] 
        for j in sub.iloc[:,5]:
            res = np.where(symptoms == j)
            df2.iloc[i, (res[0]+2)] = 1
    else:
        no_match.append(i)

df2 = df2.drop(no_match)
```
